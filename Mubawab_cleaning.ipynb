### Import Libraries

```python
import pandas as pd 
```

```python
data = pd.read_excel('MubawabMa.xlsx')
data.head()
```

- Notre DataFrame compte 1419 lignes, réparties sur 12 colonnes

```python
data.shape
```

## Suppression des colonnes

```python
#Supprimer colonne URL
data.drop(['URL'], axis =1, inplace = True)
#Supprimer colonne Caracteristiques_supplementaires
data.drop(['Caracteristiques_supplementaires'], axis =1, inplace = True)
#Supprimer colonne Etage
data.drop(['Etage'], axis =1, inplace = True)

```

```python
data.shape
```

## Supprimer les éléments indésirables d’une colonne

```python
#Colonnes Price
data['Price'] = data.Price.str.replace("€","0")
data['Price'] = data.Price.str.replace("DH","")
data['Price'] = data.Price.str.replace("EUR","0")
data['Price'] = data.Price.str.replace(" ","")
data['Price'] = data.Price.str.replace("\n","")
data['Price'] = data.Price.str.replace("\t","")
data['Price'] = data.Price.str.replace("\xa0","")
data['Price'] = data.Price.str.replace("parjour","")
data['Price'] = data.Price.str.replace("parsemaine","")


data['Price'] = data.Price.str.strip()


#Localisation
data['Localisation'] = data.Localisation.str.replace('\n','')
data['Localisation'] = data.Localisation.str.replace('\t',' ')

# Size
data['Size'] = data.Size.str.replace('m²','')
data['Size'] = data.Size.str.replace('\n','')
data['Size'] = data.Size.str.replace('\t','')

# Nb_pieces
data['Nb_pieces'] = data.Nb_pieces.str.replace(' Pièces','')
data['Nb_pieces'] = data.Nb_pieces.str.replace(' Pièce','')
data['Nb_pieces'] = data.Nb_pieces.str.replace('\n','')
data['Nb_pieces'] = data.Nb_pieces.str.replace('\t',' ')
# Nb_chambre
data['Nb_chambre'] = data.Nb_chambre.str.replace(' Chambres','')
data['Nb_chambre'] = data.Nb_chambre.str.replace(' Chambre','')
#data['Nb_chambre'] = data.Nb_chambre.str.replace('\n','')
#data['Nb_chambre'] = data.Nb_chambre.str.replace('\t',' ')
# Nb_Salles_bain
data['Nb_Salles_bain'] = data.Nb_Salles_bain.str.replace(' Salles de bains','')
data['Nb_Salles_bain'] = data.Nb_Salles_bain.str.replace(' Salle de bain','')
data['Nb_Salles_bain'] = data.Nb_Salles_bain.str.replace('En cours de construction','0')

# Etat
data['Etat'] = data.Etat.str.replace('\n','')
data['Etat'] = data.Etat.str.replace('\t',' ')
# Old
data['Old'] = data.Old.str.replace('\n','')
data['Old'] = data.Old.str.replace('\t',' ')
#Titre
data['Titre'] = data.Titre.str.replace('\n','')
data['Titre'] = data.Titre.str.replace('\t',' ')

```

```python
data.shape
```

### Suppression des valeurs manquants

```python
data.isnull().sum()

```

```python
data = data[data['Price'].notna()]
data = data[data['Size'].notna()]
data.tail(40)
```

```python
data.shape
```

```python
try:
    data['Price']= data["Price"].astype(float)
    data['Nb_pieces']= data["Nb_pieces"].astype(float)
    data['Nb_chambre']= data["Nb_chambre"].astype(float)
    data['Nb_Salles_bain']= data["Nb_Salles_bain"].astype(float)
    data['Size']= data["Size"].astype(float)
except ValueError as e:
    print(e)

    
    
```

```python

#Nous transformons toutes ces données en nombre avec .apply()

data = data.apply(pd.to_numeric, errors='ignore')
data.info()
```

### Exporter le fichier nettoyé avec .to_excel()

```python
data.to_excel('MubawabMa_clean.xlsx', index=False, encoding='utf-8')
```
